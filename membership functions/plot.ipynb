```python
import matplotlib.pyplot as plt

def plot_from_file(filename, label):
    x, y = [], []
    with open(filename, "r") as f:
        for line in f:
            xi, yi = map(float, line.split())
            x.append(xi)
            y.append(yi)
    plt.plot(x, y, label=label)

# Plot all
plot_from_file("triangular.txt", "Triangular")
plot_from_file("trapezoidal.txt", "Trapezoidal")
plot_from_file("gaussian.txt", "Gaussian")
plot_from_file("sigmoidal.txt", "Sigmoidal")
plot_from_file("singleton.txt", "Singleton")

plt.legend()
plt.xlabel("x")
plt.ylabel("Membership value (x)")
plt.title("Fuzzy Membership Functions")
plt.grid(True)
plt.show()
```
